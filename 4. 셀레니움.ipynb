```python
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from urllib.request import Request, urlopen
```

```python
driver = webdriver.Chrome()
url = "https://pixabay.com/ko/images/search/michael%20jackson/"
driver.maximize_window()
driver.get(url)
```

```python
image_xpath = "/html/body/div[1]/div[1]/div/div[2]/div[3]/div/div/div[1]/div[3]/div/a/img"
wait = WebDriverWait(driver, 10) # ÏµúÎåÄ 10Ï¥àÍπåÏßÄ Í∏∞Îã§Î¶º 
image_element = wait.until(EC.presence_of_element_located((By.XPATH, image_xpath)))
```

```python
image_url = image_element.get_attribute("src")
image_byte = Request(image_url, headers={'User-Agent':
'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/143.0.0.0 Safari/537.36'})
image_byte
```

```python
with open('mj.jpg', 'wb') as f:
    f.write(urlopen(image_byte).read())
    print('ÌååÏùºÏ†ÄÏû• ÏôÑÎ£å!!')

driver.quit()
```

```python
def crawl_yanolja_reviews(name, url, max_scroll = 3):
    review_list = []
    driver = webdriver.Chrome()
    driver.get(url)
    time.sleep(3)

    last_height = driver.execute_script("return document.body.scrollHeight")#JS Î¨∏Î≤ï
    # print(last_height)
    same_count = 0
    scrolls_time = 0

    while True:
        scrolls_time += 1 
        driver.execute_script("window.scrollTo(0, document.body.scrollHeight)")
        time.sleep(2)
        new_height = driver.execute_script("return document.body.scrollHeight")

        if new_height == last_height:
            same_count += 1
        else: 
            same_count = 0 
            last_height = new_height
        
        if same_count >= 3: 
            print("Îçî Ïù¥ÏÉÅ ÏÉàÎ°úÏö¥ Î¶¨Î∑∞Í∞Ä ÏóÜÏäµÎãàÎã§")
            break

        if scrolls_time >= max_scroll:
            print('ÏµúÎåÄ Ïä§ÌÅ¨Î°§ ÌöüÏàò ÎèÑÎã¨. Ï¢ÖÎ£åÌï©ÎãàÎã§.')
            break

        #/html/body/div[1]/section/div/div[5]/div[1]
#/html/body/div[1]/section/div/div[5]/div[1]/div[2]/div/div[3]/div[1]/p

    containers = driver.find_elements(By.XPATH, "/html/body/div[1]/section/div/div[5]/div")
    # print('Ï¥ù Î¶¨Î∑∞ Í∞úÏàò:' , len(containers))

    for c in containers:
        try:
            # raw_text = c.text.strip()
            #/html/body/div[1]/section/div/div[5]/div[34]/div[2]/div/div[3]/div[1]/p
            #/html/body/div[1]/section/div/div[5]/div[35]/div[2]/div/div[3]/div[1]/p
            date_text = c.find_element(By.XPATH, "./div[2]/div/div[1]/div[1]/div[2]/p").text
            comment_text = c.find_element(By.XPATH, "./div[2]/div/div[3]/div[1]/p").text
        except: 
            date_text=""
            comment_text=""

        review_list.append({
            "review": comment_text,
            "date": date_text
        })
            # parts = raw_text.split("\n")
            # print(parts)

            # # date = parts[1] if len(parts)
            # # room = parts[3]
            # # review = parts[4]

            # # print(date)
            # # print(room)
            # # print(review)
            # # print("-" * 30)
            
            # # print(raw_text)
            # # print("-" * 30)
        # except:
        #     raw_text=""

    driver.quit()
    return review_list
```
